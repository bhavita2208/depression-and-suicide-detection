```python
%%writefile gui.py
import streamlit as st
import numpy as np
import nltk
nltk.download('punkt')
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.models import load_model
from keras.preprocessing.text import Tokenizer
import io
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
gauth = GoogleAuth()           
drive = GoogleDrive(gauth) 



#Loading the Model
model = load_model('/content/drive/My Drive/Colab Notebooks/Suicide and Depression/saved_model/model_dnn_pre_text_stem.h5', compile=False)




st.title("Suicide and Depresion Detection")
test = st.text_area("Add Text Here...")

submit_button = st.button("Submit")
if submit_button:
    labels=["Depresion","Suicide"]
       
    
    tokenizer = Tokenizer(num_words=10000, oov_token='')
    tokenizer.fit_on_texts(test)

    word_index = tokenizer.word_index

    x_train_sequ = tokenizer.texts_to_sequences(nltk.word_tokenize(test))
    training_padded = pad_sequences(x_train_sequ, maxlen=500, padding='post', truncating='post')
    training_padded = np.array(training_padded)
    y_pred = model.predict(training_padded)
    y_pred = np.argmax(y_pred, axis=1)
    st.text(y_pred)
    st.success("Result: " + labels[y_pred[0]]) 


st.markdown('Sample Data')
st.text('Took all my pills, be dead soon, bye bye everyone')
st.text('medication working damage life depression')
st.text('problem issues nothing live want kill justifia')


```
